```python
import pandas as pd
from pathlib import Path
```

```python
read_file = pd.read_excel (".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.xlsx")
# Write the dataframe object
# into csv file
read_file.to_csv (".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.csv", 
                  index = None,header=False)
# read csv file and convert 
# into a dataframe object
cleaned_county_df = pd.DataFrame(pd.read_csv(".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.csv"))
```

```python
cleaned_county_df.rename(columns = {"State and County of Geography A":"State, County"}, inplace = True)
cleaned_county_df.rename(columns = {"Net Migration from Geography B to Geography A1":"Net Migration"}, inplace = True)
cleaned_county_df.head()
```

```python
file_path = ".\Resources\County_population_totals_2010_2019_co-est2019-alldata.csv"
total_county_df = pd.read_csv(file_path)
total_county_df.rename(columns = {"State and County name":"State, County"}, inplace = True)
total_county_df.head()
```

```python
joined_df=pd.merge(cleaned_county_df,total_county_df,on="State, County")
joined_df.head()
```

```python
table_df=joined_df[["State Name of Geography A","State, County", "State/U.S. Island Area/Foreign Region of Geography B",
                         "Net_Migration_from_Geography_B_to_Geography_A1",
                        "POPESTIMATE2015"]]
table_df.head()
```

```python
table_df.rename(columns = {"Net_Migration_from_Geography_B_to_Geography_A1":"Net Migration"}, inplace = True)
table_df.rename(columns = {"State Name of Geography A":"State"}, inplace = True)
table_df.rename(columns = {"State/U.S. Island Area/Foreign Region of Geography B":"Migrated Area"}, inplace = True)
table_df.rename(columns = {"POPESTIMATE2015":"Pop. Estimate"}, inplace = True)
table_df.head()
```

```python
table_df['Net Migration'] = table_df['Net Migration'].fillna(0)
```

```python
table_df.dtypes
```

```python
table_df['Net Migration']=pd.to_numeric(table_df['Net Migration'], errors='coerce').fillna(0).astype(int)
```

```python
table_df.dtypes
```

```python
table_df["Percent Change in Pop."]=table_df["Net Migration"]/table_df["Pop. Estimate"]
table_df.head()
```

```python
pivot_table = pd.pivot_table(table_df, index=["State", "State, County","Migrated Area"], 
                        values=["Net Migration","Pop. Estimate","Percent Change in Pop."])
pivot_table
```
