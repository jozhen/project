```python
import pandas as pd
from pathlib import Path
```

```python
file_path = ".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.csv"
cleaned_county_df = pd.read_csv(file_path)
cleaned_county_df.rename(columns = {"State and County of Geography A":"State and County name"}, inplace = True)
cleaned_county_df.rename(columns = {"Net Migration from Geography B to Geography A1":"Net_Migration_from_Geography_B_to_Geography_A1"}, inplace = True)
cleaned_county_df.head()
```

```python
file_path = ".\Resources\County_population_totals_2010_2019_co-est2019-alldata.csv"
total_county_df = pd.read_csv(file_path)
total_county_df.head()
```

```python
joined_df=pd.merge(cleaned_county_df,total_county_df,on="State and County name")
joined_df.head()
```

```python
table_df=joined_df[["State and County name", "State/U.S. Island Area/Foreign Region of Geography B",
                         "Net_Migration_from_Geography_B_to_Geography_A1",
                        "POPESTIMATE2015"]]
table_df.head()
```

```python
table_df['Net_Migration_from_Geography_B_to_Geography_A1'] = table_df['Net_Migration_from_Geography_B_to_Geography_A1'].fillna(0)
```

```python
#table_df.to_csv (".\Resources\merge_pop_dataframe.csv", index=True, header=True)
```

```python
table_df.dtypes
```

```python
table_df['Net_Migration_from_Geography_B_to_Geography_A1']=pd.to_numeric(table_df['Net_Migration_from_Geography_B_to_Geography_A1'], errors='coerce').fillna(0).astype(int)
```

```python
table_df.dtypes
```

```python
table_df["Percent_Change_in_Pop"]=table_df["Net_Migration_from_Geography_B_to_Geography_A1"]/table_df["POPESTIMATE2015"]
table_df
```

```python
pivot_table = pd.pivot_table(table_df, index=["State and County name","State/U.S. Island Area/Foreign Region of Geography B"], 
                        values=["Net_Migration_from_Geography_B_to_Geography_A1","POPESTIMATE2015","Percent_Change_in_Pop"])
pivot_table
```
